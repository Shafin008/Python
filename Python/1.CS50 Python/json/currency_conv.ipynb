## loading json foramted string

```python
import json
people_string = '''{"people":[{ "name": "John Smith",
            "phone": "615-555-7164",
            "emails": ["johnsmith@bogusmail.com", "john,smith@work-place.com"],
            "has_license": false},

            {"name": "Jane Doe",
            "phone": "560-555-5153",
            "emails": null,
            "has_license": true}]}'''

## Loading the json file as python object
data = json.loads(people_string)
print(data,'\n')
## after loading the type of the json file will be a dictionary
print(type(data))
```

***JSON -> Python***

object -> dict

array -> list

string -> str

number(int) -> int

number(real) -> float

true -> True

false -> False

null -> None

## dumping json formated string

```python
## we'll create a json file from this existing file
## let's just say we'll delete the phone numbers

for person in data["people"]:
    del person['phone']
```

```python
## data will have no phone numbers
data
```

```python
## we'll dump our file, indent and sort it
new_str = json.dumps(data, indent = 2, sort_keys=True)
print(new_str)
```

```python
%%writefile currency_converter.py
import json
from urllib.request import urlopen

def main():
    print("Please enter currencies in small letter abbreviation. Like, american dollar will be usd")
    print(converter())
    
def converter():
    with urlopen("http://www.floatrates.com/daily/usd.json") as response:
        source = response.read()
        source = source.decode("utf-8")  
        data = json.loads(source)
    
    for curr in data:
        name = data[curr]['name']
        price = data[curr]['rate']
        
    currency = input('Currency you want to convert: ')
    if currency == 'usd':
        curr_name = "USD"
        curr_rate = 1.0
    else:    
        curr_name = data[currency]['name']
        curr_rate = data[currency]['rate']
    
    converted_curr = input('Convert to: ')
        
    if converted_curr == "usd":
        conv_curr_name = "USD"
        converted_curr_rate = 1.0
        
    else:
        conv_curr_name = data[converted_curr]['name']
        converted_curr_rate = data[converted_curr]['rate']
    
    currency_amt = input("Currency amount: ")
    currency_amt = float(currency_amt)
    return f"{currency_amt:.2f} {curr_name} is {currency_amt*(converted_curr_rate/curr_rate):.2f} {conv_curr_name}."

if __name__ == "__main__":
    main()
```

```python
%%writefile currency_coverter_sys.py
import sys
import json
from urllib.request import urlopen


def main():
    print("Converting currency....")
    print(converter())
    
def converter():
    with urlopen("http://www.floatrates.com/daily/usd.json") as response:
        source = response.read()
        source = source.decode("utf-8")  
        data = json.loads(source)
    
    for curr in data:
        name = data[curr]['name']
        price = data[curr]['rate']
        
    currency = sys.argv[1]
    if currency == 'usd':
        curr_name = "USD"
        curr_rate = 1.0
    else:    
        curr_name = data[currency]['name']
        curr_rate = data[currency]['rate']
    
    converted_curr = sys.argv[2]
        
    if converted_curr == "usd":
        conv_curr_name = "USD"
        converted_curr_rate = 1.0
        
    else:
        conv_curr_name = data[converted_curr]['name']
        converted_curr_rate = data[converted_curr]['rate']
    
    currency_amt = sys.argv[3]
    currency_amt = float(currency_amt)
    return f"{currency_amt:.2f} {curr_name} is {currency_amt*(converted_curr_rate/curr_rate):.2f} {conv_curr_name}."

if __name__ == "__main__":
    main()
```
