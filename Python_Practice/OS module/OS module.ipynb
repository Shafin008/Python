```python
## os module allows us to interact with underlying operating system
import os
```

```python
## If we use dir(any_module_we_are_working) then it'll show us all the attributes and methods of that module
## Let's use this with os module
print(dir(os))
```

```python
## Current directory
print(os.getcwd())
```

```python
## changing the current directory
os.chdir('/Volumes/GoogleDrive/Other computers/My Computer/CODE/')

## Current directory
print(os.getcwd())
```

```python
## To see the file and folders of the current directory

print(os.listdir())
```

```python
## Making a directory or folder

os.mkdir('Nothing')
print(os.listdir())
```

```python
## But for making directories inside a directory we need to use another method

os.makedirs('Nothing_2/sub_dir_1')

## changing the current directory
os.chdir('/Volumes/GoogleDrive/Other computers/My Computer/CODE/Nothing_2')

print(os.listdir())
```

```python
## changing the current directory
os.chdir('/Volumes/GoogleDrive/Other computers/My Computer/CODE/')

## removing a folder
os.removedirs('Nothing')
```

```python
print(os.listdir())
```

```python
## removing a folder
os.removedirs('Nothing_2/sub_dir_1')
print(os.listdir())
```

```python
## renaming a file or folder

os.rename('710 711', '710_711')

print(os.listdir())
```

```python
## to find any information about a folder or a file

os.stat('Scipy/')
```

```python
## st_size attribute tells us the file size in bytes
os.stat("Scipy/").st_size
```

```python
from datetime import datetime
```

```python
## st_mtime shows us the last modification time of that specific file and folder
os.stat("Scipy/").st_mtime
```

```python
## to get a clear view of modification time we need to use datetime module

mod_time = os.stat("Scipy/").st_mtime
print(datetime.fromtimestamp(mod_time))
```

```python
### to check all of the files and folders in directory by using os.walk()

for dirpath, dirnames, filenames in os.walk('/Volumes/GoogleDrive/Other computers/My Computer/CODE/'):
    print("Current path:", dirpath, '\n')
    print("Directories:", dirnames, '\n')
    print("files:", filenames)
    print()
    print('******************************************************')
    print()
```

```python
print(os.environ.get("HOME"))
```

```python
## if we want to create a file on this home directory we can use os.path() module

file_path = os.path.join(os.environ.get("HOME"), 'test.txt')
print(file_path)
```

```python
## manipulating a path
## lets check with a fake path

print(os.path.dirname('tmp/test.txt')) # this file path doesn't exist, just for example
print(os.path.basename('tmp/test.txt')) # it'll print the base file name
print(os.path.split('tmp/test.txt')) # it'll split the main directory and sub directory
print(os.path.exists('tmp/test.txt')) # it'll print out if the folder exists or not
```
